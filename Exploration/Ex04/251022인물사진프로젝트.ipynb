```python
import cv2
import numpy as np
import torch
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet101
import matplotlib.pyplot as plt

print(cv2.__version__)
```

```python
import os
import requests

cat_img_path = "/content/sample_data/cat.jpg"  # 고양이 이미지
sand_img_path = "/content/sample_data/sand.jpg"  # 사막 배경 이미지


cat_img = cv2.imread(cat_img_path)
sand_img = cv2.imread(sand_img_path)

cat_img = cv2.cvtColor(cat_img, cv2.COLOR_BGR2RGB)
sand_img = cv2.cvtColor(sand_img, cv2.COLOR_BGR2RGB)

print(f"원본 고양이 이미지 크기: {cat_img.shape}")
print(f"배경(사막) 이미지 크기: {sand_img.shape}")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cat_img)

plt.subplot(1, 2, 2)
plt.imshow(sand_img)

plt.show()
```

```python
model = deeplabv3_resnet101(pretrained=True).eval()

transform = T.Compose([
    T.ToPILImage(),
    T.Resize((520, 520)),  # 모델 입력 크기
    T.ToTensor(),
])

input_tensor = transform(cat_img).unsqueeze(0)

with torch.no_grad():
    output = model(input_tensor)["out"][0]
    output_predictions = output.argmax(0).byte().cpu().numpy()

print(f"추론 마스크 크기 (Before Resize): {output_predictions.shape}")
```

```python
#pascalvoc 데이터의 라벨종류
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
len(LABEL_NAMES)
```

```python
# 마스크를 원본 크기로 Resize
output_predictions_resized = cv2.resize(output_predictions, (cat_img.shape[1], cat_img.shape[0]), interpolation=cv2.INTER_NEAREST)

print(f"추론 마스크 크기 (After Resize): {output_predictions_resized.shape}")
```

```python
# 고양이를 제대로 구분하였네요.
unique_classes = np.unique(output_predictions_resized)
print(f"예측된 클래스 ID: {unique_classes}")
```

```python
# 고양이의 누끼를 따 보았습니다.
target_class_id = unique_classes[-1]

seg_map = (output_predictions_resized == target_class_id)
img_mask1 = seg_map.astype(np.uint8) * 255


plt.imshow(img_mask1, cmap='gray')
plt.show()
```

```python
# 배경을 고양이 이미지 크기에 맞춤
sand_img_resized = cv2.resize(sand_img, (cat_img.shape[1], cat_img.shape[0]))

# 고양이 부분만 남기고 배경 적용
img_mask_color = cv2.cvtColor(img_mask1, cv2.COLOR_GRAY2BGR)  # 3채널 변환
result1_img = np.where(img_mask_color == 255, cat_img, sand_img_resized)  # 마스크 기반 합성

plt.imshow(result1_img)
plt.show()
# 고양이와 사막배경을 합성해봤습니다. 결과가 뭔가 이상합니다. 배경 아웃포커싱이 되지 않으니 원근감이 맞지 않아서 매우 이상합니다.
```

```python
# 모레 이미지에 블러(아웃포커싱)를 넣어봅어보려고 합니다. 고양이 이미지는 현재 포커싱되어 있는 상태입니다
img_orig_blur = cv2.blur(sand_img, (13, 13))

# plt.imshow(): 저장된 데이터를 이미지의 형식으로 표시한다.
# cv2.cvtColor(입력 이미지, 색상 변환 코드): 입력 이미지의 색상 채널을 변경
# cv2.COLOR_BGR2RGB: 원본이 BGR 순서로 픽셀을 읽다보니
# 이미지 색상 채널을 변경해야함 (BGR 형식을 RGB 형식으로 변경)
plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 블러된 배경을 고양이 이미지 크기에 맞춤
sand_img_resized = cv2.resize(img_orig_blur, (cat_img.shape[1], cat_img.shape[0]))

# 고양이 부분만 남기고 배경 적용
img_mask_color = cv2.cvtColor(img_mask1, cv2.COLOR_GRAY2BGR)  # 3채널 변환
result2_img = np.where(img_mask_color == 255, cat_img, sand_img_resized)  # 마스크 기반 합성

plt.imshow(result2_img)
plt.show()
# 아직도 결과가 뭔가 이상합니다. 아웃포커싱이 들어가서 원근감이 일정부분 구현되었으나 뭔가 다소 어색합니다. 고양이의 사이즈를 줄여보는 것이 현실적인 그림이 될 듯합니다.
```

**개선방향 분석**

고양이 사진과 배경이 비율이 맞지 않으므로 사진이 비현실적으로 보임. 다음 작업에서 배경과 고양이 사진의 비율을 조정하겠음





```python
# 다음은 고양이 사진의 비율을 조정해서 다시 합성하기 위한 작업입니다.
# Define the desired size for the larger background. 배경 이미지를 고양이 이미지의 3배로 키웁니다.
new_bg_width = cat_img.shape[1] * 3
new_bg_height = cat_img.shape[0] * 3

# Resize the blurred background to the new larger size
sand_img_resized_large = cv2.resize(img_orig_blur, (new_bg_width, new_bg_height))

# Create a blank canvas the size of the new large background
result2_img = np.copy(sand_img_resized_large)

# Get the dimensions of the original cat image
cat_height_orig, cat_width_orig, _ = cat_img.shape

# Calculate the position to place the cat image (e.g., centered) -- 위치를 조정했습니다. 아웃 포커싱 효과를 생각하면 고양이 이미지의 위치는 맨아래로 적합합니다.
# You might want to adjust these based on where you want the cat to appear
start_y = (new_bg_height - cat_height_orig)
start_x = (new_bg_width - cat_width_orig) // 2
end_y = start_y + cat_height_orig
end_x = start_x + cat_width_orig

# Ensure the slice indices are within the bounds of the larger background
start_y = max(0, start_y)
start_x = max(0, start_x)
end_y = min(new_bg_height, end_y)
end_x = min(new_bg_width, end_x)

# Create a 3-channel mask for the original cat size
img_mask1_color = cv2.cvtColor(img_mask1, cv2.COLOR_GRAY2BGR)

# Apply the mask to the relevant region of the larger background
# We need to apply the mask to the part of the larger background where the cat will be placed
# And then put the masked cat image on top
masked_cat = np.where(img_mask1_color == 255, cat_img, np.zeros_like(cat_img))

# Get the region of the larger background where the cat will be placed
bg_region = result2_img[start_y:end_y, start_x:end_x]

# Create the inverse mask
inverse_mask = cv2.bitwise_not(img_mask1)
inverse_mask_color = cv2.cvtColor(inverse_mask, cv2.COLOR_GRAY2BGR)

# Apply the inverse mask to the background region
masked_bg_region = np.where(inverse_mask_color == 255, bg_region, np.zeros_like(bg_region))


# Combine the masked background region and the masked cat
result2_img[start_y:end_y, start_x:end_x] = masked_bg_region + masked_cat


plt.imshow(result2_img)
plt.show()
```

**개선방향 분석**

1. 여러 인물 사진에서 비교적 목적에 맞게 segmentation이 이루어지나 부분적으로 미흡한 부분이 있다. 이는 보다 발전된 모델을 사용하는 것이 필요할 것으로 보이며 바로 적용하기에는 현실적인 한계가 있어 보임.

2. 배경 인물 이미지 사이즈를 줄이고, 멀리 배치하며, 아웃포커싱 하는 방식으로 변환하는 것이 보다 자연스러운 이미지를 만들 수 있는 효과적인 솔류션으로 판단됨. 본 프로젝트는 2번 안을 선택하였음.

```python
# 고양이 사진과 같은 방식으로 비율을 조정해보겠음
# Define the desired size for the larger background. 배경 이미지를 사람들 이미지의 3배로 키웁니다.
new_bg_width2 = people_img.shape[1] * 3
new_bg_height2 = people_img.shape[0] * 3

# Resize the background to the new larger size
inwater_img_resized_large = cv2.resize(inwater_img, (new_bg_width2, new_bg_height2))

# Create a blank canvas the size of the new large background
result5_img = np.copy(inwater_img_resized_large)

# Get the dimensions of the original people image
people_height_orig, people_width_orig, _ = people_img.shape

# Calculate the position to place the people image (e.g., centered)
# You might want to adjust these based on where you want the people to appear
start_y = (new_bg_height2 - people_height_orig) // 2
start_x = (new_bg_width2 - people_width_orig) // 2
end_y = start_y + people_height_orig
end_x = start_x + people_width_orig

# Ensure the slice indices are within the bounds of the larger background
start_y = max(0, start_y)
start_x = max(0, start_x)
end_y = min(new_bg_height2, end_y)
end_x = min(new_bg_width2, end_x)

# Create a 3-channel mask for the original people size
img_mask2_color = cv2.cvtColor(img_mask2, cv2.COLOR_GRAY2BGR)

# Apply the mask to the relevant region of the larger background
# We need to apply the mask to the part of the larger background where the people will be placed
# And then put the masked people image on top
masked_people = np.where(img_mask2_color == 255, img_orig_blur2, np.zeros_like(people_img))

# Get the region of the larger background where the people will be placed
bg_region = result5_img[start_y:end_y, start_x:end_x]

# Create the inverse mask
inverse_mask2 = cv2.bitwise_not(img_mask2)
inverse_mask2_color = cv2.cvtColor(inverse_mask2, cv2.COLOR_GRAY2BGR)

# Apply the inverse mask to the background region
masked_bg_region2 = np.where(inverse_mask2_color == 255, bg_region, np.zeros_like(bg_region))


# Combine the masked background region and the masked people
result5_img[start_y:end_y, start_x:end_x] = masked_bg_region2 + masked_people


plt.imshow(result5_img)
plt.show()
```
